```python
# Install required library
!pip install pandas openpyxl -q

import pandas as pd

# Upload your Excel file in Colab before running
input_file = "BMP_ADASYN_ALL_Results.xlsx"

# Output file name
output_file = "BMP_ADASYN_ALL_Results_With_STD.xlsx"

# Read all sheets
all_sheets = pd.read_excel(input_file, sheet_name=None)

# Create output Excel writer
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:

    # Process every sheet
    for sheet_name, df in all_sheets.items():

        # Remove old Std row if already present
        df = df[df['Fold'] != 'Std']

        # Remove Mean row for STD calculation
        temp_df = df[df['Fold'] != 'Mean'].copy()

        # Select numeric columns
        numeric_cols = temp_df.columns[1:]

        # Convert to numeric
        temp_df[numeric_cols] = temp_df[numeric_cols].apply(
            pd.to_numeric,
            errors='coerce'
        )

        # Calculate standard deviation
        std_values = temp_df[numeric_cols].std(ddof=1)

        # Create Std row
        std_row = ['Std'] + list(std_values.values)

        # Append Std row
        df.loc[len(df)] = std_row

        # Save updated sheet
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print(" Standard deviation calculated successfully!")
print(f" Output file saved as: {output_file}")
```
